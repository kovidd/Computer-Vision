```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

def unsharp_mask(image, kernel_size=(5, 5), sigma=1.0, amount=1.25, threshold=0):
    """Return a sharpened version of the image, using an unsharp mask."""
    
    blurred = cv.GaussianBlur(image, kernel_size, sigma)
    sharpened = float(amount + 1) * image - float(amount) * blurred
#     cv.imwrite('cat_1.png', sharpened_image)
    
    sharpened = np.maximum(sharpened, np.zeros(sharpened.shape))
#     cv.imwrite('cat_2.png', sharpened_image)
    
    sharpened = np.minimum(sharpened, 255 * np.ones(sharpened.shape))
#     cv.imwrite('cat_3.png', sharpened_image)
    
    sharpened = sharpened.round().astype(np.uint8)
#     cv.imwrite('cat_4.png', sharpened_image)
    if threshold > 0:
        low_contrast_mask = np.absolute(image - blurred) < threshold
        np.copyto(sharpened, image, where=low_contrast_mask)
    return sharpened


image = cv.imread('cat.png')
for i in range(5):
    sharpened_image = unsharp_mask(image, (5,5), i)
    plt.imshow(sharpened_image, cmap = 'gray')
    plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('cat.png')
kernel = np.array([[-1,-1,-1], 
                   [-1, 9,-1],
                   [-1,-1,-1]])
sharpened = cv2.filter2D(image, -1, kernel) # applying the sharpening kernel to the input image & displaying it.
plt.imshow(sharpened, cmap = 'gray')
plt.show()
# cv2.imwrite('cat_0_0.png', sharpened)
```
